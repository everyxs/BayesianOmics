```python
import pickle

with open ('hierarchy.save', 'rb') as fp:
    itemlist = pickle.load(fp)

hierarchy  = [x for x in itemlist if "subsystem" in str(x[0])]
mapping = [x for x in itemlist if "subsystem" not in str(x[0])]
```

```python
hierarchy
```

```python
import ddot
from ddot import Ontology
# Construct ontology
ont = Ontology(hierarchy, mapping)

# Prints a summary of the ontology's structure
print(ont)
```

```python
ont.to_table('toy_ontology.txt')
```

```python
# Reconstruct the ontology from the table
ont2 = Ontology.from_table('toy_ontology.txt')
ont2
```

```python
G = ont.to_networkx()
print(G.nodes())
print(G.edges())
```

```python
import networkx as nx
#import matplotlib.pyplot as plt

graphs = [c for c in nx.connected_components(G.to_undirected())]
print(len(graphs))
```

```python
sorted([n for n in G.out_degree()], reverse=True)  # degree sequence
```

```python
outDegrees = G.out_degree()
for node, oDegree in outDegrees.items():    # for name, age in dictionary.iteritems():  (for Python 2.x)
    if oDegree == 3:
        print(node)
```

```python
inDegrees = G.in_degree()
for node, iDegree in inDegrees.items():    # for name, age in dictionary.iteritems():  (for Python 2.x)
    if iDegree == 1:
        print(node)
```
