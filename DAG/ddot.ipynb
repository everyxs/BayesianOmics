```python
Gorig = nx.read_edgelist("data/resnik.txt", delimiter='\t' , nodetype=str,  data=(('weight',float),))
with open('hierarchyQ3clixo.save', 'wb') as fp:
    pickle.dump(Gorig.edges(), fp)
```

```python
from networkx import relabel

sim = nx.adjacency_matrix(Gorig).todense()
Gnew = relabel.convert_node_labels_to_integers(Gorig, label_attribute = "term") #initilaization
terms = nx.get_node_attributes(Gnew, 'term')
terms
```

```python
ont1 = Ontology.run_clixo(sim, alpha=0.0, beta=1.0, square=True, square_names=terms)
print(ont1)
ont1.to_table('toy_ontology3clixo.txt')
```
