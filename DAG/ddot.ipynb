```python
import pickle

with open ('hierarchy.save', 'rb') as fp:
    itemlist = pickle.load(fp)

hierarchy  = [x for x in itemlist if "subsystem" in str(x[0])]
mapping = [x for x in itemlist if "subsystem" not in str(x[0])]
```

```python
hierarchy
```

```python
import ddot
from ddot import Ontology
# Construct ontology
ont = Ontology(hierarchy, mapping)

# Prints a summary of the ontology's structure
print(ont)
```

```python
ont.to_table('toy_ontology.txt')
```

```python
# Reconstruct the ontology from the table
ont2 = Ontology.from_table('toy_ontology.txt')
ont2
```
