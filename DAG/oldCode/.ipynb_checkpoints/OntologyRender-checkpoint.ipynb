```python
G0 = nx.read_edgelist("data/pearl2015.txt", delimiter='\t' , nodetype=str,  data=(('label',str),),create_using=nx.DiGraph())
A0 = nx.adjacency_matrix(G0)
type(A0)
plt.spy(A0, markersize=5)
```

```python
import community
G0U = G0.to_undirected()
partition = community.best_partition(G0U)
#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(G0U)
count = 0.
for com in set(partition.values()) :
    count += 1.
    list_nodes = [nodes for nodes in partition.keys()
        if partition[nodes] == com]
    nx.draw_networkx_nodes(G0U, pos, list_nodes, node_size = 20,node_color = str(count / size))
nx.draw_networkx_edges(G0U, pos, alpha=0.5)
plt.show()
```

```python
T0 = nx.minimum_spanning_tree(G0U)
nodes = T0.nodes()
E = set(T0.edges())  # optimization
Ed = [e for e in G0.edges() if e in E or reversed(e) in E]
T = nx.DiGraph()
T.add_nodes_from(nodes)
T.add_edges_from(Ed)
```

```python
del(Z0)
```

```python
#T = nx.convert_node_labels_to_integers(G0)
d =  nx.to_dict_of_lists(T)
leaves = set( n for n in nodes if T.out_degree(n) == 0 )
inner_nodes = [ n for n in nodes if T.out_degree(n) > 0 ]
# Compute the size of each subtree
subtree = dict( (n, [n]) for n in leaves )
for u in inner_nodes:
    children = set()
    node_list = list(d[u])
    while len(node_list) > 0:
        v = node_list.pop(0)
        children.add( v )
        node_list += d[v]
    subtree[u] = sorted(children & leaves)

inner_nodes.sort(key=lambda n: len(subtree[n])) # <-- order inner nodes ascending by subtree size, root is last
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Construct the linkage matrix
leaves = sorted(leaves)
index  = dict( (tuple([n]), i) for i, n in enumerate(leaves) )
Z0 = []
k = len(leaves)
for i, n in enumerate(inner_nodes):
    children = d[n]
    if children:
        x = children[0]
        for y in children[1:]:
            z = tuple(sorted(subtree[x] + subtree[y]))
            #print(y)
            u, v = index[tuple(subtree[x])], index[tuple(subtree[y])]            
            dist = float(len(subtree[n]))
            Z0.append([u, v, dist, len(z)]) # <-- float is required by the dendrogram function
            index[z] = k
            subtree[z] = list(z)
            x = z
            k += 1

# Visualize
if Z0:
    dendrogram(Z0, labels=leaves)
    plt.show()
```
